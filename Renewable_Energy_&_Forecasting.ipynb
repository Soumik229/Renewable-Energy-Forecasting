# **Regression Part**

Regression is a supervised machine learning technique used to predict a continuous value (numeric output).

Example with our dataset:
- Predicting Renewable Energy Production (GWh) from Installed Capacity, Investments, Climate conditions, Policies etc.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import RobustScaler
```

###**Spliting Training and Testing Set**


```python
# Split into features and target
X = data.drop("Production (GWh)", axis=1)  # Features
y = data["Production (GWh)"]
X.shape , y.shape
```

```python
# Convert categorical columns into numbers
X = pd.get_dummies(X, drop_first=True)
```

```python
# Split the data into traning and testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Check the shape of train and test datasets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape : {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape : {y_test.shape}")
```

```python
X_train[0:10]
```

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

###**Linear Regression**

In machine learning, linear regression is a supervised learning algorithm that models the relationship between a dependent variable (the target) and one or more independent variables (features) by finding a linear equation that best fits the data.


```python
from sklearn.linear_model import LinearRegression
```

```python
# Create a linear Regression model
lr = LinearRegression()
lr
```

```python
# Train the model on the traning set (This is how the model will learn.).
lr.fit(X_train, y_train)
```

```python
# Predictions on the testing set.
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
# Evaluation the performence of the model.

r2 = r2_score(y_test, y_pred)
adj_r2 = 1 - (1-r2) * (len(y_test)-1) / (len(y_test)-X_test.shape[1]-1)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
intercept = lr.intercept_

print("Linear Regression Results:")
print("-----------------------------------")
print("R² Score:", r2)
print("Adjusted R²:", adj_r2)
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("Intercept:", intercept)
```

```python
# Plot the pradicted vs. actuals value
plt.figure(figsize=(7,5))
plt.scatter(y_test, y_pred, color="blue", alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "r--")
plt.xlabel("Actual Production (GWh)")
plt.ylabel("Predicted Production (GWh)")
plt.title("Actual vs Predicted - Linear Regression")
plt.show()
```

### **Logistic Regression**

Logistic regression is a statistical method used for classification problems, specifically those with a binary outcome (two possible results)

###**Decision Tree**
Decision tree is a type of supervised learning algorithm that is mostly used in classification problems. It works for both categorical and continuous input and output variables.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
```

```python
# 'Production (GWh)' is our dependent variable (y)
y = df['Production (GWh)']
X = df.drop('Production (GWh)', axis=1)

# Convert categorical columns into numbers using one-hot encoding
X = pd.get_dummies(X, columns=['Country', 'Energy Type'], drop_first=True)

# Convert continuous target into categories (Low, Medium, High)
y_binned = pd.qcut(y, q=3, labels=['Low', 'Medium', 'High'])

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_binned, test_size=0.2, random_state=42)

# Train Decision Tree Classifier
dt_clf = DecisionTreeClassifier(random_state=42, max_depth=4)
dt_clf.fit(X_train, y_train)
```

```python
# Plot the tree
plt.figure(figsize=(20,10))
plot_tree(dt_clf,
          feature_names=X_train.columns,
          class_names=dt_clf.classes_,
          filled=True,
          rounded=True,
          fontsize=8)
plt.show()
```

Forecasting = predicting future values of a variable based on its historical trends.
Example:

- Predicting Renewable Energy Production (GWh) for the next 10 years.

- Estimating Investments in Renewables in future.

ARIMA stands for Autoregressive Integrated Moving Average. It is a statistical method used for time series analysis and forecasting.

### **Random Forest**
In machine learning, a Random Forest is an ensemble learning method that constructs a "forest" of multiple decision trees during training and outputs the mode of the classes (for classification problems) or the mean prediction (for regression problems) of the individual trees.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Create model
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=0,
    max_depth=None,
    n_jobs=-1
)

# Split the data into training and testing sets again to ensure y_train is numerical
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train model
rf_model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = rf_model.predict(X_test)

# Evaluation
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Random Forest Performance:")
print(f"  MSE  = {mse:.2f}")
print(f"  RMSE = {rmse:.2f}")
print(f"  MAE  = {mae:.2f}")
print(f"  R²   = {r2:.4f}")
```

```python
# 1. Predicted vs Actual
plt.figure(figsize=(12,6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # 45° line
plt.xlabel("Actual Production (GWh)")
plt.ylabel("Predicted Production (GWh)")
plt.title("Random Forest: Predicted vs Actual")
plt.show()
```

```python
# 2. Residuals Distribution
residuals = y_test - y_pred
plt.figure(figsize=(10,6))
sns.histplot(residuals, kde=True, bins=30, color="purple")
plt.axvline(0, color="red", linestyle="--")
plt.xlabel("Residuals (Actual - Predicted)")
plt.title("Residuals Distribution")
plt.show()
```

```python
# Get feature importance
importances = rf_model.feature_importances_
features = X.columns

# Create a DataFrame for better handling
feat_imp = pd.DataFrame({"Feature": features, "Importance": importances})

# Sort by importance and select top 10
top_features = feat_imp.sort_values(by="Importance", ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
plt.barh(top_features["Feature"], top_features["Importance"], color="green")
plt.gca().invert_yaxis()  # Highest at top
plt.xlabel("Importance")
plt.title("Top 10 Feature Importances in Random Forest")
plt.show()
```

### **Comparison Between models.**
- Here we are going for the Comparison between The Linear Regrassion model and Random Forest model and also Comparison between the Decision Tree model and Logistic Regression model

```python
# Comparison result for Regression Models

reg_models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(n_estimators=100, random_state=0)
}

reg_results = {}

for name, model in reg_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    reg_results[name] = [rmse, mae, r2]

reg_results_df = pd.DataFrame(reg_results, index=["RMSE", "MAE", "R²"]).T

print("\nRegression Model Comparison Results:")
print("-----------------------------------")
print(reg_results_df)
```

```python
# Comparison Results for Classification Model
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42, max_depth=4),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42)
}

results = {}

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

clf_results_df = pd.DataFrame.from_dict(results, orient="index", columns=["Accuracy"])

print("\nClassification Model Comparison Results:")
print("-----------------------------------")
print(clf_results_df)
```
